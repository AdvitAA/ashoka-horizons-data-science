```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
starbucks_df = pd.read_csv('starbucks.csv')

# Remove leading/trailing spaces from column names to prevent errors
starbucks_df.columns = starbucks_df.columns.str.strip()

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=starbucks_df, x='Calories', y='Sugars (g)')
plt.title('Relationship between Calories and Sugar Content in Starbucks Menu Items')
plt.xlabel('Calories')
plt.ylabel('Sugar (g)')
plt.grid(True)
plt.show()
```

```python
#The scatter plot shows how calories and sugar are connected for Starbucks menu items. From this, we can try to see if there's a pattern ‚Äî like if more calories usually means more sugar, or if there‚Äôs not really a clear link between them.

#If we notice a positive trend, that means the stuff with higher calories also usually has more sugar, which makes sense for things like sweet drinks or desserts. But if the pattern looks weird or scattered, then maybe some things have lots of calories but not much sugar, like maybe protein-heavy items, or some things have tons of sugar but not as many calories. This helps in figuring out what kinds of items are what, nutritionally speaking.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
netflix_df = pd.read_csv('netflix_titles.csv')

# Count titles per release year
yearly_counts = netflix_df['release_year'].value_counts().sort_index()

# Create labeled line plot
plt.figure(figsize=(12, 6))
plt.plot(yearly_counts.index, yearly_counts.values, marker='o')
plt.title('Number of Netflix Titles Released Each Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
netflix_df = pd.read_csv('netflix_titles.csv')

# Count titles per release year by type
movie_counts = netflix_df[netflix_df['type'] == 'Movie']['release_year'].value_counts().sort_index()
tv_counts = netflix_df[netflix_df['type'] == 'TV Show']['release_year'].value_counts().sort_index()

# Create line plot with separate lines
plt.figure(figsize=(12, 6))
plt.plot(movie_counts.index, movie_counts.values, marker='o', label='Movies')
plt.plot(tv_counts.index, tv_counts.values, marker='s', label='TV Shows')
plt.title('Number of Netflix Titles Released Each Year by Type')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
pima_df = pd.read_csv('diabetes.csv')

# Select 5 numerical features
selected_features = ['Glucose', 'BloodPressure', 'BMI', 'Age', 'Outcome']

# Calculate correlation matrix
correlation_matrix = pima_df[selected_features].corr()

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Selected PIMA Features')
plt.show()
```

```python
#When looking at the correlation heatmap, we can spot some feature pairs that have very high or very low values. If two features have a correlation close to 1 or -1, like more than 0.7 or less than -0.7, that usually means they change together in a strong way. For example, if Glucose and Outcome have a high positive correlation, it means that when glucose goes up, the chance of diabetes also tends to go up.

#If two features have correlation near zero, like between -0.3 and 0.3, then they don‚Äôt really affect each other much. For example, if BloodPressure and Age show low correlation, that means those two things don‚Äôt really move together in this data.

#This matters because if features are very highly correlated, they might be giving the same kind of information, and it could be too much of the same thing for a model. But when features are not strongly related, each one can give its own kind of information. Correlations with the Outcome column are especially useful because they help show which variables are probably good at telling us if someone has diabetes or not.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
pima_df = pd.read_csv('diabetes.csv')
starbucks_df = pd.read_csv('starbucks.csv')

# Clean column names for consistency
starbucks_df.columns = starbucks_df.columns.str.strip()

# Create 2x2 subplot grid
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Top-left: Histogram of BloodPressure from PIMA dataset
axes[0, 0].hist(pima_df['BloodPressure'], bins=20, color='lightblue', edgecolor='black')
axes[0, 0].set_title('Distribution of Blood Pressure')
axes[0, 0].set_xlabel('Blood Pressure')
axes[0, 0].set_ylabel('Frequency')

# Top-right: Scatter plot of Calories vs Protein from Starbucks dataset
axes[0, 1].scatter(starbucks_df['Calories'], starbucks_df['Protein (g)'], alpha=0.7)
axes[0, 1].set_title('Calories vs Protein Content')
axes[0, 1].set_xlabel('Calories')
axes[0, 1].set_ylabel('Protein (g)')

# Bottom-left: Count plot of Beverage category from Starbucks dataset
beverage_counts = starbucks_df['Beverage_category'].value_counts()
axes[1, 0].bar(range(len(beverage_counts)), beverage_counts.values)
axes[1, 0].set_title('Distribution of Beverage Categories')
axes[1, 0].set_xlabel('Beverage Category')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_xticks(range(len(beverage_counts)))
axes[1, 0].set_xticklabels(beverage_counts.index, rotation=45, ha='right')

# Bottom-right: Box plot of Age grouped by Outcome from PIMA dataset
outcome_0 = pima_df[pima_df['Outcome'] == 0]['Age']
outcome_1 = pima_df[pima_df['Outcome'] == 1]['Age']
axes[1, 1].boxplot([outcome_0, outcome_1], labels=['No Diabetes', 'Diabetes'])
axes[1, 1].set_title('Age Distribution by Diabetes Outcome')
axes[1, 1].set_xlabel('Outcome')
axes[1, 1].set_ylabel('Age')

# Add overall title and adjust layout
fig.suptitle('Multi-Dataset Analysis Dashboard', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
